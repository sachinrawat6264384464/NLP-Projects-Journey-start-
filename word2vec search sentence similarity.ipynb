✅ Text preprocessing using of NLP(natural preprocess language)

Typical cleaning steps:

1 remove HTML tags

2 lowercase

3 remove punctuation

4 remove stopwords optional in word2vec this process is not necessory of this ❌

5 lemmatization

```python
from bs4 import BeautifulSoup

data["clean_text"] = data["headline_text"].apply(
    lambda x: BeautifulSoup(str(x), "html.parser").get_text()
)
```

```python

def to_lowercase(text):
    
    if isinstance(text, str):
        text = text.lower()
    
    return text
lower=data["clean_text"].apply(to_lowercase)
print("convert into lowercase  successfully")
```

```python
import re
def remove_punctuation(text):
    
    punctuation = r"[~!@#$%^&*\-?;:\"'+=_`>,<./(){}\[\]|]"
    
    if isinstance(text, str):
        text = re.sub(punctuation, "", text)
    
    return text
re_pun=lower.apply(remove_punctuation)
print("remove punchautions successfully")
```

right here we we use nltk for convert sentances into tokensization form

```python
import nltk
nltk.download("stopwords")
nltk.download("punkt")


nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(tokens):
    
    if isinstance(tokens, list):
        return [lemmatizer.lemmatize(word) for word in tokens]
    
    return tokens
lemin=re_pun.apply(lemmatize_words)
print(lemin)
```

```python
from nltk.tokenize import word_tokenize
def tokenize_text(text):
    if isinstance(text, str):
        return word_tokenize(text)
    return text
new_tokens=lemin.apply(tokenize_text)
print(new_tokens)
```

know we apply th model word2vec 

```python
from gensim.models import Word2Vec

model1 = Word2Vec(
    new_tokens,
    vector_size=100,
    window=5,
    min_count=2,
    workers=1
)
```
