```python
import re

def remove_punctuation(text):
    
    punctuation = r"[~!@#$%^&*\-?;:\"'+=_`>,<./(){}\[\]|]"
    
    if isinstance(text, str):
        text = re.sub(punctuation, "", text)
    
    return text


y_cle = y_clea.apply(remove_punctuation)
x_cle = x_clea.apply(remove_punctuation)

print(y_cle)
print(x_cle)
```

```python
import re

def remove_url(text):
    
    url_pattern = r'https?://\S+|www\.\S+'
    
    if isinstance(text, str):
        text = re.sub(url_pattern, '', text)
    
    return text
```

```python
x_cl = x_cle.apply(remove_url)
y_cl = y_cle.apply(remove_url)
print(x_cl)
print(y_cl)
```

```python
def to_lowercase(text):
    
    if isinstance(text, str):
        text = text.lower()
    
    return text
```

```python
x_c = x_cl.apply(to_lowercase)
y_c = y_cl.apply(to_lowercase)
print(x_c)
print(y_c)
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
def tokenize_text(text):
    if isinstance(text, str):
        return word_tokenize(text)
    return text


x_tokens = x_c.apply(tokenize_text)
y_tokens = y_c.apply(tokenize_text)

print(x_tokens.head())
print(y_tokens.head())
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    
    if isinstance(tokens, list):
        filtered = [word for word in tokens if word not in stop_words]
        return filtered
    
    return tokens
x_stopw = x_tokens.apply(remove_stopwords)
y_stopw = y_tokens.apply(remove_stopwords)

print(x_stopw.head())
print(y_stopw.head())
```

```python
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

def lemmatize_words(tokens):
    
    if isinstance(tokens, list):
        return [lemmatizer.lemmatize(word) for word in tokens]
    
    return tokens
```

```python
x_stem = x_stopw.apply(lemmatize_words)
y_stem = y_stopw.apply(lemmatize_words)
print(x_stem)
print(y_stem)
```
